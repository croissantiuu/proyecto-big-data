```python
import numpy as np

# Función para extraer la fecha del diccionario
def extraer_fecha(diccionario):
    # Comprueba si el valor es NaN
    if pd.isna(diccionario):
        return ''
    # Convierte la cadena del diccionario en un diccionario real
    diccionario = ast.literal_eval(diccionario)
    # Comprueba si la fecha está vacía
    if diccionario['date'] == '':
        return ''
    else:
        # Devuelve la fecha
        return diccionario['date']

# Aplica la función a la columna 'Fecha de Lanzamiento'
df['Fecha'] = df['Fecha de Lanzamiento'].apply(extraer_fecha)

# Muestra el DataFrame
print(df.head())
```

```python
from datetime import datetime

def formatearFecha(fecha_str):
    if pd.isnull(fecha_str) or not isinstance(fecha_str, str):
        return fecha_str
    else:
        formatos = ["%d %b, %Y", "%d-%b-%Y", "%d/%m/%Y", "%Y-%m-%d"]  # Agrega aquí otros formatos que quieras probar
        for formato in formatos:
            try:
                fecha_obj = datetime.strptime(fecha_str, formato)
                return fecha_obj
            except ValueError:
                continue
        print(f"No se pudo parsear la fecha: {fecha_str}")
        return -1  # Devuelve la cadena original si no se pudo parsear

# Asegúrate de que df es tu DataFrame
df['Fecha'] = df['Fecha'].apply(formatearFecha)
```

```python
import requests
def diasDesde(fecha):
    if pd.isnull(fecha) or fecha == -1.0:
        return -1  # Devuelve None si fecha no es un objeto datetime
    else:
        hoy = datetime.today().date()
        return (hoy - fecha.date()).days

# Asegúrate de que df es tu DataFrame
df['Dias_Pasados'] = df['Fecha'].apply(diasDesde)
```

```python
mediaDeDias = int(df['Dias_Pasados'].mean())
print(mediaDeDias)
```

```python
def cleanMinusOne(posibleMinusOne):
    if posibleMinusOne == -1:
        return mediaDeDias
    else:
        return posibleMinusOne
df['Dias_Pasados'] = df['Dias_Pasados'].apply(cleanMinusOne)
```
